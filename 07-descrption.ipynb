```python
import pyreadstat
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams["font.sans-serif"]=["SimHei"]#设置字体
```

```python
df=pd.read_spss(R'identity.sav')
```

```python
#无需类别变量的描述统计
df['政治面貌'].value_counts(normalize=True)
```

```python
#生成x,y轴方向的坐标
x=df['政治面貌'].value_counts().index
y=df['政治面貌'].value_counts(normalize=True).values*100
#创建图
fig,ax=plt.subplots()
#绘制柱状图
rects1=ax.bar(x,y)
#设置x轴变量名称
ax.set_xlabel('政治面貌')
#设置y轴最大值
ax.set_ylim(ymax=100)
#在柱上方显示对应的值
ax.bar_label(rects1,fmt="%.1f",padding=3)
#显示图形
plt.show()
```

```python
import matplotlib.pyplot as plt  
  
def draw_pie_chart(labels, sizes):  
    plt.pie(sizes, labels=labels, autopct='%1.1f%%')  
    plt.show()   
labels = ['苹果', '香蕉', '橙子', '葡萄']  
sizes = [30, 25, 20, 25]    
draw_pie_chart(labels, sizes)
```

```python
import matplotlib.pyplot as plt
def draw_pie_chart(数据表, 变量):
    plt.pie(变量, 数据表=数据表, autopct='%1.1f%%', startangle=90)
    plt.axis('equal')  
    plt.show()
```
